```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import requests
import time
from config import stock_key
```

```python
function = 'TIME_SERIES_DAILY'
symbol = ['AA','KALU','CENX', 'ACH', 'ARNC']
apikey = stock_key

url = 'https://www.alphavantage.co/'
```

```python
date_final = []
       
start_date = pd.datetime(2018, 3 ,1)
end_date = pd.datetime(2018, 3, 31)
range_one = pd.bdate_range(start_date, end_date)
dates = (range_one.strftime('%Y-%m-%d')).tolist()
dates.remove('2018-03-30')

for stock in symbol:
    time.sleep(80)
    for x in dates:'5'
    query_url = requests.get(url + 'query?function=' + function + '&' + 'symbol=' + stock + '&' + 'apikey=' 
                    + apikey).json()
    date = query_url['Time Series (Daily)'][x]['4. close']
    
    date_final.append({'Stock':  stock, 'Date' : x, 'Close_Price' : date})
        
    pd.DataFrame(date_final)
```
