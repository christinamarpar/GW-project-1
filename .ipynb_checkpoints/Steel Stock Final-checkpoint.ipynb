```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import requests
import time
from config import stock_key
```

```python
function = 'TIME_SERIES_DAILY'
stock_symbol = ['NUE', 'X', 'STLD', 'AKS', 'CMC']
apikey = stock_key

url = 'https://www.alphavantage.co/'
```

```python
stock_data = []
       
dates = ['2018-03-01', '2018-03-08']    

for stock in stock_symbol:
    time.sleep(60)
    for x in dates:
        query_url = requests.get(url + 'query?function=' + function + '&' + 'symbol=' + stock + '&' + 'apikey=' + apikey).json()
        open_price = query_url['Time Series (Daily)'][x]['1. open']
        close_price = query_url['Time Series (Daily)'][x]['4. close']   
        low_day = query_url['Time Series (Daily)'][x]['3. low']   
        volume_day = query_url['Time Series (Daily)'][x]['5. volume']                                       
                                
                                         
        stock_data.append({'Stock':  stock, 'Date' : x, 'Open' : open_price, 'Close' : close_price, 'Low': low_day, 'Volume' : volume_day})
        data_final = pd.DataFrame(stock_data, dtype = 'float')
```
